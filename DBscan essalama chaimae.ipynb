```python
X = df.iloc[:, 1:14].values
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
```

```python
principalComponents
```

```python
PCA_dataset = pd.DataFrame(data = principalComponents, columns = ['Component 1', 'Component 2'] )
PCA_dataset.head()
```

```python
component1 = PCA_dataset['Component 1']
component2 = PCA_dataset['Component 2']
```

```python
plt.figure()
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2 Component PCA')
plt.scatter(PCA_dataset['Component 1'], PCA_dataset['Component 2'], c = 'red', s=50)
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(principalComponents, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Compounds')
plt.ylabel('Euclidean distances')
plt.show()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 0.25, min_samples=4, metric='euclidean').\
fit(principalComponents)
model
```

```python
from collections import Counter
outliers_df = PCA_dataset[model.labels_ == -1]
clusters_df = PCA_dataset[model.labels_ != -1]
colors = model.labels_
colors_clusters = colors[colors != -1]
colors_outliers = 'black'
clusters = Counter(model.labels_)
print(clusters)
print(PCA_dataset[model.labels_ == -1].head())
print("Number of clusters = {}".format(len(clusters)-1))
```

```python
fig = plt.figure()

ax = fig.add_axes([1,1,1,1])

ax.scatter(clusters_df['Component 1'], clusters_df['Component 2'], 
                      c = colors_clusters, edgecolors='black', s=50)

ax.scatter(clusters_df['Component 1'], clusters_df['Component 2'], 
                      c = colors_outliers, edgecolors='black', s=50)

ax.set_xlabel('Principal Component 1', family='Arial', fontsize=9)
ax.set_ylabel('Principal Component 2', family='Arial', fontsize=9)

plt.title('Clustered Compounds', family='Arial', fontsize=12)

plt.grid(which = 'major', color = '#cccccc', alpha=0.45)

plt.show()
```
